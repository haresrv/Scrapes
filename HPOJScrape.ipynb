```python
with requests.Session() as s:
    url = 'http://hpoj.cb.amrita.edu:8000/users'
    load_users=s.get(url,headers=headers)
    usertab = BeautifulSoup(load_users.content,'html.parser')
```

```python
import re
```

```python
usertab.findAll("table")[0].findAll("th",{"class":"header"})[3]
```

```python
rank=[]
username=[]
point=[]
problems=[]
```

```python
for i in usertab.findAll("table")[0].select("tr[id*=user]"):
    rank.append(i.findAll("td")[0].text.strip())
    username.append(i.findAll("td")[2].text)
    point.append(i.findAll("td")[3].text.strip())
    problems.append(i.findAll("td")[4].text.strip())
```

```python
cols=['Username','Point','Problems']
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data={'Username':username,'Point':point,'Problems':problems},columns=cols,index=rank)
```
